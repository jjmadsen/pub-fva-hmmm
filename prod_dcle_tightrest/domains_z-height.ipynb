```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest/domains_z-height.csv', sep=',')
df2=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest.2/domains_z-height.csv', sep=',')
df3=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest.3/domains_z-height.csv', sep=',')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, 1, gridspec_kw={'height_ratios': [1, 1, 1]}, figsize=(8, 6))
ax1.plot(df1.Frame,df1.A1_z,color='darkorange',label='A1')
ax1.plot(df1.Frame,df1.A2_z,color='gold',label='A2')
ax1.plot(df1.Frame,df1.A3_z,color='green',label='A3')
ax1.plot(df1.Frame,df1.C1_z,color='blue',label='C1')
ax1.plot(df1.Frame,df1.C2_z,color='red',label='C2')
ax1.plot(df1.Frame,df1.FVa,color='black',label='FVa')
ax2.plot(df1.Frame,df2.A1_z,color='darkorange',label='A1')
ax2.plot(df1.Frame,df2.A2_z,color='gold',label='A2')
ax2.plot(df1.Frame,df2.A3_z,color='green',label='A3')
ax2.plot(df1.Frame,df2.C1_z,color='blue',label='C1')
ax2.plot(df1.Frame,df2.C2_z,color='red',label='C2')
ax2.plot(df1.Frame,df2.FVa,color='black',label='FVa')
ax3.plot(df1.Frame,df3.A1_z,color='darkorange',label='A1')
ax3.plot(df1.Frame,df3.A2_z,color='gold',label='A2')
ax3.plot(df1.Frame,df3.A3_z,color='green',label='A3')
ax3.plot(df1.Frame,df3.C1_z,color='blue',label='C1')
ax3.plot(df1.Frame,df3.C2_z,color='red',label='C2')
ax3.plot(df1.Frame,df3.FVa,color='black',label='FVa')
ax2.set_ylabel("Domain or protein relative height, $z$ [$\AA$]",size=15)
ax3.set_xlabel("Simulation time [ns]",size=15)
ax1.set_ylim([0,120])
ax2.set_ylim([0,120])
ax3.set_ylim([0,120])
ax1.set_xticklabels([0,0,100,200,300,400,500])
ax2.set_xticklabels([0,0,100,200,300,400,500])
ax3.set_xticklabels([0,0,100,200,300,400,500])
ax1.set_yticks([0,20,40,60,80,100,120])
ax2.set_yticks([0,20,40,60,80,100,120])
ax3.set_yticks([0,20,40,60,80,100,120])
ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
```
