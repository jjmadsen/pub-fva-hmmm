```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest/tilt_angle.csv', sep=',')
df2=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest.2/tilt_angle.csv', sep=',')
df3=pd.read_csv('/Users/jespermadsen/Projects/FV-HMMM/output_from_RRA/production_chlen7/openmm_dcle_tightrest.3/tilt_angle.csv', sep=',')
```

```python
df3
```

```python
fig, ax = plt.subplots()
ax.plot(df1.Frame,df1.tilt_angle,color='blue',label='Memb #1')
ax.plot(df2.Frame,df2.tilt_angle,color='red',label='Memb #2')
ax.plot(df3.Frame,df3.tilt_angle,color='green',label='Memb #3')
ax.legend(loc="upper left")
ax.set_ylim([0,75])
ax.set_ylabel(r"Molecular tilt angle, $\alpha$ [deg]",size=15)
ax.set_xlabel("Simulation time [ns]",size=15)
ax.set_xticklabels([0,0,100,200,300,400,500])
```

```python
fig.savefig("/Users/jespermadsen/Projects/FV-HMMM/figures/fva_tilt/tilt_reps.pdf")
```

```python
bins=np.histogram(np.hstack((df1.tilt_angle,df2.tilt_angle)), bins=100)[1]
plt.hist(df1.tilt_angle[3000:5000],bins,color='blue')
plt.hist(df2.tilt_angle[3000:5000],bins,color='red')
plt.hist(df3.tilt_angle[3000:5000],bins,color='green')
```
